```python
# Question ( mapping + data-type casting)¶
#4.what is the mean stop_duration?
```

```python
data.head(2)
```

```python
# Convert the 'stop_duration' column to numerical values and fill missing values with the mean
data['stop_duration'] = pd.to_numeric(data['stop_duration'], errors='coerce')
mean_stop_duration = data['stop_duration'].mean()
data['stop_duration'].fillna(mean_stop_duration, inplace=True)
```

```python
# Create a mapping for stop durations
duration_mapping = {
    '0-15 Min': 7.5,
    '16-30 Min': 23,
    '30+ Min': 45
}

# Map the string values to numerical values
data['stop_duration'] = data['stop_duration'].map(duration_mapping)
```

```python
from scipy.stats import zscore

# Calculate the z-scores for the 'stop_duration' column
z_scores = zscore(data['stop_duration'])

# Define a threshold for outliers (e.g., z_score > 3 or z_score < -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Replace outliers with the mean value
data['stop_duration'][outliers] = mean_stop_duration
```

```python
# Calculate the mean stop duration
mean_stop_duration = data['stop_duration'].mean()
```
